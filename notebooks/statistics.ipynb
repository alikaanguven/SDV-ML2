```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import rv_continuous

# Define a linear PDF: p(x) ∝ x over [0, 1]
class linear_pdf(rv_continuous):
    def _pdf(self, x):
        return (3*x**2 + 2*x + 2) / 4 # normalized on [0,1] (since ∫0^1 2x dx = 1)

# Instantiate the distribution
linear_dist = linear_pdf(a=0, b=1, name='linear')

# Sample values
samples = linear_dist.rvs(size=100000, random_state=42)

# Plot histogram vs theoretical PDF
x = np.linspace(0, 1, 200)
plt.hist(samples, bins=100, density=True, alpha=0.6, label='samples')
plt.plot(x, linear_dist.pdf(x), 'r-', lw=2, label='theoretical PDF')
plt.xlabel('x')
plt.ylabel('density')
plt.legend()
plt.title("Linear probability distribution (p(x) ∝ x)")
plt.show()
```
